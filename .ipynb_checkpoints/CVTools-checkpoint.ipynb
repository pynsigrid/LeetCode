## IOU

```python
# 1å¯¹1çš„iouè®¡ç®—
def iou_count(bboxA, bboxB):
    # äº¤å‰åŒºåŸŸçš„åæ ‡
    xa = max(bboxA[0], bboxB[0])
    ya = max(bboxA[1], bboxB[1])
    xb = min(bboxA[2], bboxB[2])
    yb = min(bboxA[3], bboxB[3])

    # äº¤å‰é¢ç§¯è®¡ç®—
    inter = max(0, xb-xa+1)*max(0, yb-ya+1)  # ğŸ“ä¿è¯ä¸¤ä¸ªå®Œå…¨é‡å bboxçš„iouä¸º1

    # bboxAå’ŒbboxBçš„é¢ç§¯
    areaA = (bboxA[2]-bboxA[0]+1)*(bboxA[3]-bboxA[1]+1)
    areaB = (bboxB[2]-bboxB[0]+1)*(bboxB[3]-bboxB[1]+1)
    
    # è®¡ç®—iou
    iou = inter/float(areaA+areaB-inter)

    return iou
```

```python
bboxA, bboxB = [1,1,4,4], [2,2,8,8]
print(iou_count(bboxA, bboxB))
```

```python
def nms(dets, thre):
    x1, y1, x2, y2, scores = dets[:, 0], dets[:, 1], dets[:, 2], dets[:, 3], dets[:, 4]
    areas = abs(x2-x1+1)*abs(y2-y1+1)
    order = scores.argsort()[::-1]
    
    res_idx = []
    while order.size>0:
        target_idx = order[0]
        res_idx.append(target_idx)
        
        xa = np.maximum(x1[target_idx], x1[order[1:]])  # np.maximumé€ä½æ¯”è¾ƒ
        ya = np.maximum(y1[target_idx], y1[order[1:]])
        xb = np.minimum(x2[target_idx], x2[order[1:]])
        yb = np.minimum(y2[target_idx], y2[order[1:]])
        
        inter = np.maximum(0, xb-xa+1)*np.maximum(0, yb-ya+1)
        iou = inter / (areas[target_idx]+areas[order[1:]]-inter)
        
        skip_idx = np.where(iou<=thre)[0]
        print(skip_idx, order, order[skip_idx+1])
        order = order[skip_idx+1]
    
    return res_idx
```

```python
dets = np.array([[30, 20, 230, 200, 1], 
                 [50, 50, 260, 220, 0.9],
                 [210, 30, 420, 5, 0.8],
                 [430, 280, 460, 360, 0.7]])
thresh = 0.35
keep_dets = nms(dets, thresh)
```

```python
# python3
import numpy as np

def py_nms(dets, thresh):
    # 1. å¯¹æ‰€æœ‰detsä¸­çš„ç‚¹å¤„ç†ï¼Œè®¡ç®—æ¯ä¸€ä¸ªç‚¹çš„é¢ç§¯
    # x1ã€y1ã€x2ã€y2ã€ä»¥åŠscoreèµ‹å€¼
    x1 = dets[:, 0]
    y1 = dets[:, 1]
    x2 = dets[:, 2]
    y2 = dets[:, 3]
    scores = dets[:, 4]

    # æ¯ä¸€ä¸ªå€™é€‰æ¡†çš„é¢ç§¯
    areas = (x2 - x1 + 1) * (y2 - y1 + 1)
    #orderæ˜¯æŒ‰ç…§scoreé™åºæ’åºçš„
    order = scores.argsort()[::-1]

    keep = []
    while order.size > 0:
        i = order[0]
        keep.append(i)
        #è®¡ç®—å½“å‰æ¦‚ç‡æœ€å¤§çŸ©å½¢æ¡†ä¸å…¶ä»–çŸ©å½¢æ¡†çš„ç›¸äº¤æ¡†çš„åæ ‡ï¼Œä¼šç”¨åˆ°numpyçš„broadcastæœºåˆ¶ï¼Œå¾—åˆ°çš„æ˜¯å‘é‡
        xx1 = np.maximum(x1[i], x1[order[1:]])
        yy1 = np.maximum(y1[i], y1[order[1:]])
        xx2 = np.minimum(x2[i], x2[order[1:]])
        yy2 = np.minimum(y2[i], y2[order[1:]])

        #è®¡ç®—ç›¸äº¤æ¡†çš„é¢ç§¯,æ³¨æ„çŸ©å½¢æ¡†ä¸ç›¸äº¤æ—¶wæˆ–hç®—å‡ºæ¥ä¼šæ˜¯è´Ÿæ•°ï¼Œç”¨0ä»£æ›¿
        w = np.maximum(0.0, xx2 - xx1 + 1)
        h = np.maximum(0.0, yy2 - yy1 + 1)
        inter = w * h
        #è®¡ç®—é‡å åº¦IOUï¼šé‡å é¢ç§¯/ï¼ˆé¢ç§¯1+é¢ç§¯2-é‡å é¢ç§¯ï¼‰
        ovr = inter / (areas[i] + areas[order[1:]] - inter)

        #æ‰¾åˆ°é‡å åº¦ä¸é«˜äºé˜ˆå€¼çš„çŸ©å½¢æ¡†ç´¢å¼•
        inds = np.where(ovr <= thresh)[0]
        #å°†orderåºåˆ—æ›´æ–°ï¼Œç”±äºå‰é¢å¾—åˆ°çš„çŸ©å½¢æ¡†ç´¢å¼•è¦æ¯”çŸ©å½¢æ¡†åœ¨åŸorderåºåˆ—ä¸­çš„ç´¢å¼•å°1ï¼Œæ‰€ä»¥è¦æŠŠè¿™ä¸ª1åŠ å›æ¥
        order = order[inds + 1]
    return keep

# test
if __name__ == "__main__":
    dets = np.array([[30, 20, 230, 200, 1], 
                     [50, 50, 260, 220, 0.9],
                     [210, 30, 420, 5, 0.8],
                     [430, 280, 460, 360, 0.7]])
    thresh = 0.35
    keep_dets = py_nms(dets, thresh)
    print(keep_dets)
    print(dets[keep_dets])

```
