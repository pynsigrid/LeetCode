```python
class SingleLinkList(object):
    """å•é“¾è¡¨"""

    def __init__(self):
        self._head = None

    def is_empty(self):
        """åˆ¤æ–­é“¾è¡¨æ˜¯å¦ä¸ºç©º"""
        return self._head is None

    def length(self):
        """é“¾è¡¨é•¿åº¦"""
        # åˆå§‹æŒ‡é’ˆæŒ‡å‘head
        cur = self._head
        count = 0
        # æŒ‡é’ˆæŒ‡å‘None è¡¨ç¤ºåˆ°è¾¾å°¾éƒ¨
        while cur is not None:
            count += 1
            # æŒ‡é’ˆä¸‹ç§»
            cur = cur.next
        return count

    def items(self):
        """éå†é“¾è¡¨"""
        # è·å–headæŒ‡é’ˆ
        cur = self._head
        # å¾ªç¯éå†
        while cur is not None:
            # è¿”å›ç”Ÿæˆå™¨
            yield cur.val
            # æŒ‡é’ˆä¸‹ç§»
            cur = cur.next

    def add(self, item):
        """å‘é“¾è¡¨å¤´éƒ¨æ·»åŠ å…ƒç´ """
        node = Node(item)
        # æ–°ç»“ç‚¹æŒ‡é’ˆæŒ‡å‘åŸå¤´éƒ¨ç»“ç‚¹
        node.next = self._head
        # å¤´éƒ¨ç»“ç‚¹æŒ‡é’ˆä¿®æ”¹ä¸ºæ–°ç»“ç‚¹
        self._head = node

    def append(self, item):
        """å°¾éƒ¨æ·»åŠ å…ƒç´ """
        node = ListNode(item)
        # å…ˆåˆ¤æ–­æ˜¯å¦ä¸ºç©ºé“¾è¡¨
        if self.is_empty():
            # ç©ºé“¾è¡¨ï¼Œ_head æŒ‡å‘æ–°ç»“ç‚¹
            self._head = node
        else:
            # ä¸æ˜¯ç©ºé“¾è¡¨ï¼Œåˆ™æ‰¾åˆ°å°¾éƒ¨ï¼Œå°†å°¾éƒ¨nextç»“ç‚¹æŒ‡å‘æ–°ç»“ç‚¹
            cur = self._head
            while cur.next is not None:
                cur = cur.next  # âš ï¸æ¯æ¬¡ç”Ÿæˆæ–°é“¾è¡¨åï¼Œ_headéƒ½é‡æ–°æŒ‡å‘å¤´éƒ¨ï¼Œå› æ­¤éœ€è¦éå†åˆ°é“¾å°¾
            cur.next = node

    def insert(self, index, item):
        """æŒ‡å®šä½ç½®æ’å…¥å…ƒç´ """
        # æŒ‡å®šä½ç½®åœ¨ç¬¬ä¸€ä¸ªå…ƒç´ ä¹‹å‰ï¼Œåœ¨å¤´éƒ¨æ’å…¥
        if index <= 0:
            self.add(item)
        # æŒ‡å®šä½ç½®è¶…è¿‡å°¾éƒ¨ï¼Œåœ¨å°¾éƒ¨æ’å…¥
        elif index > (self.length() - 1):
            self.append(item)
        else:
            # åˆ›å»ºå…ƒç´ ç»“ç‚¹
            node = ListNode(item)
            cur = self._head
            # å¾ªç¯åˆ°éœ€è¦æ’å…¥çš„ä½ç½®
            for i in range(index - 1):
                cur = cur.next
            node.next = cur.next
            cur.next = node

    def remove(self, item):
        """åˆ é™¤èŠ‚ç‚¹"""
        cur = self._head
        pre = None
        while cur is not None:
            # æ‰¾åˆ°æŒ‡å®šå…ƒç´ 
            if cur.val == item:
                # å¦‚æœç¬¬ä¸€ä¸ªå°±æ˜¯åˆ é™¤çš„èŠ‚ç‚¹
                if not pre:
                    # å°†å¤´æŒ‡é’ˆæŒ‡å‘å¤´èŠ‚ç‚¹çš„åä¸€ä¸ªèŠ‚ç‚¹
                    self._head = cur.next
                else:
                    # å°†åˆ é™¤ä½ç½®å‰ä¸€ä¸ªèŠ‚ç‚¹çš„nextæŒ‡å‘åˆ é™¤ä½ç½®çš„åä¸€ä¸ªèŠ‚ç‚¹
                    pre.next = cur.next
                return True
            else:
                # ç»§ç»­æŒ‰é“¾è¡¨åç§»èŠ‚ç‚¹
                pre = cur
                cur = cur.next

    def find(self, item):
        """æŸ¥æ‰¾å…ƒç´ æ˜¯å¦å­˜åœ¨"""
        return item in self.items()
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        if not head or not head.next:   #èŠ‚ç‚¹ä¸å­˜åœ¨æˆ–ä¸‹ä¸€ä¸ªèŠ‚ç‚¹ä¸å­˜åœ¨
            return False
            
        slow = head
        fast = head.next
        
        while fast and fast.next:
            if slow == fast:
                return True
            slow = slow.next
            fast = fast.next.next
            
        return False
```

## 141. ç¯å½¢é“¾è¡¨II


```python
class Solution(object):
    def detectCycle(self, head):
        slow, quick= head, head
        while quick and quick.next:
            slow = slow.next
            quick = quick.next.next
            if slow == quick:
                quick = head
                while slow != quick:
                    slow = slow.next
                    quick = quick.next
                return slow
        return None
```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
        
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if l1 is None:
            return l2
        if l2 is None:
            return l1
        if l1.val < l2.val:
            l1.next = self.mergeTwoLists(l1.next, l2)
            return l1
        else:
            l2.next = self.mergeTwoLists(l1, l2.next)
            return l2
            
```

2. ğŸ“l_headåœ¨ç¬¬ä¸€ä¸ªç»“ç‚¹æ’ä¸€ä¸ªæŒ‡é’ˆçš„æ–¹æ³•

æ—¶é—´å¤æ‚åº¦ï¼š$O(M+N)$

ç©ºé—´å¤æ‚åº¦ï¼š$O(1)$

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
        
        
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        l_head = ListNode(-1)  # ç›¸å½“äºåœ¨ç¬¬ä¸€ä¸ªç»“ç‚¹æ’ä¸€ä¸ªæŒ‡é’ˆï¼Œè¿”å›è¯¥æŒ‡é’ˆè€Œä¸æ˜¯læœ«å°¾çš„æŒ‡é’ˆ
        l = l_head
        while l1 and l2:
            if l1.val < l2.val:
                l.next = l1
                l1 = l1.next
            else:
                l.next = l2
                l2 = l2.next
            l = l.next
        l.next = l1 if l1 else l2
        return l_head.next
        
```

```python
l1, l2 = ListNode(1), ListNode(2)
l3, l4 = ListNode(3), ListNode(4)
l5, l6, l22= ListNode(10), ListNode(6), ListNode(8)
# l1
l4.next = l5
l3.next = l4
l1.next = l3
# l2
l6.next = l22
l2.next = l6

a = Solution()
res = a.mergeTwoLists(l1, l2)
res.val
```

## 160. ç›¸äº¤é“¾è¡¨

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None


class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        ha, hb = headA, headB
        while ha != hb:
            ha = ha.next if ha else headB
            hb = hb.next if hb else headA
        return ha
```

```python
l1, l2 = ListNode(3), ListNode(2)
l3, l4 = ListNode(1), ListNode(4)
l5, l6, l22= ListNode(6), ListNode(6), ListNode(4)
# l1
l4.next = l5
l3.next = l4
l1.next = l3
# l2
l2.next = l4

a = Solution()
res = a.getIntersectionNode(l1, l2)
res.val
```
